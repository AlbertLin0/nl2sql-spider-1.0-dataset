```python
import mysql.connector
import os

def create_db(database):
    # 1. 生成数据库名列表
    db_list = [f"spider1_0_{i}" for i in database]

    # 2. MySQL 连接配置（不要指定 database）
    DB_CONFIG = {
        "host": "10.16.41.190",
        "user": "root",      # 替换为您的用户名
        "password": "Ydkj.nlp@2025",  # 替换为您的密码
        "port": "12345"  
    }

    def create_databases(db_list, config):
        """
        连接到 MySQL 实例并批量创建数据库
        """
        conn = None
        cursor = None
        try:
            print("尝试连接到 MySQL 服务器...")
            conn = mysql.connector.connect(
                host=config["host"],
                user=config["user"],
                password=config["password"],
                port=config.get("port", 3306),
                autocommit=True
            )
            cursor = conn.cursor()
            print("连接成功，启用 AUTOCOMMIT 模式。")

            for db_name in db_list:
                try:
                    print(f"正在创建数据库: {db_name}...")
                    cursor.execute(
                        f"CREATE DATABASE IF NOT EXISTS `{db_name}`;"
                    )
                    print(f"✅ 数据库 '{db_name}' 创建成功（或已存在）。")

                except mysql.connector.Error as e:
                    print(f"❌ 创建数据库 '{db_name}' 失败: {e}")

        except mysql.connector.Error as e:
            print("\n❌ 致命错误：无法连接到 MySQL 服务器。")
            print(e)

        finally:
            if cursor:
                cursor.close()
            if conn:
                conn.close()
                print("\n数据库连接已关闭。")

    create_databases(db_list, DB_CONFIG)
```

# 提交脚本

```python
def split_sql_statements(sql: str):
    statements = []
    buffer = []

    in_single_quote = False
    in_double_quote = False
    in_line_comment = False
    in_block_comment = False

    i = 0
    length = len(sql)

    while i < length:
        char = sql[i]
        next_char = sql[i + 1] if i + 1 < length else ''

        # 行注释 --
        if not in_single_quote and not in_double_quote and not in_block_comment:
            if char == '-' and next_char == '-':
                in_line_comment = True
                i += 2
                continue

        # 块注释 /* */
        if not in_single_quote and not in_double_quote and not in_line_comment:
            if char == '/' and next_char == '*':
                in_block_comment = True
                i += 2
                continue

        if in_line_comment:
            if char == '\n':
                in_line_comment = False
            i += 1
            continue

        if in_block_comment:
            if char == '*' and next_char == '/':
                in_block_comment = False
                i += 2
            else:
                i += 1
            continue

        # 字符串处理
        if char == "'" and not in_double_quote:
            in_single_quote = not in_single_quote

        elif char == '"' and not in_single_quote:
            in_double_quote = not in_double_quote

        # 分号：只有在不在字符串中才切
        if char == ';' and not in_single_quote and not in_double_quote:
            statement = ''.join(buffer).strip()
            if statement:
                statements.append(statement)
            buffer = []
        else:
            buffer.append(char)

        i += 1

    # 最后一条
    last = ''.join(buffer).strip()
    if last:
        statements.append(last)

    return statements
```

```python
import mysql.connector

def commit(database):
    database = f"spider1_0_{database}"

    DB_CONFIG = {
        "host": "10.16.41.190",
        "user": "root",
        "password": "Ydkj.nlp@2025",
        "database": database,
        "port": 12345,
        "autocommit": False
    }

    SQL_FILE_PATH = f"{database}-schema.sql"

    conn = None
    cursor = None
    try:
        print(f"连接 MySQL 数据库: {database}")
        conn = mysql.connector.connect(**DB_CONFIG)
        cursor = conn.cursor()

        with open(SQL_FILE_PATH, "r", encoding="utf8") as f:
            sql_text = f.read()

        statements = split_sql_statements(sql_text)
        print(f"拆分得到 {len(statements)} 条 SQL")

        for idx, stmt in enumerate(statements, 1):
            try:
                print(f"[{idx}/{len(statements)}] 执行: {stmt[:80]}...")
                cursor.execute(stmt)
            except mysql.connector.Error as e:
                print(f"❌ SQL 执行失败（第 {idx} 条）")
                print(stmt)
                raise e

        conn.commit()
        print("✅ 所有 SQL 执行完成")

    except Exception as e:
        print(f"❌ 出错: {e}")
        if conn:
            conn.rollback()
            print("事务回滚（注意：DDL 无法回滚）")

    finally:
        if cursor:
            cursor.close()
        if conn:
            conn.close()
            print("连接关闭")
```

```python
import mysql.connector
import os

def commit2(database):
    database = f"spider1_0_{database}"

    # --- 1. MySQL 数据库连接配置 ---
    DB_CONFIG = {
        "host": "10.16.41.190",
        "database": database,
        "user": "root",              # MySQL 用户
        "password": "Ydkj.nlp@2025",
        "port": 12345,                # ⚠️ 改成你的 MySQL 端口
        "autocommit": False
    }

    # --- 2. SQL 文件路径 ---
    SQL_FILE_PATH = f"{database}-data.sql"

    def execute_sql_file():
        conn = None
        cursor = None
        try:
            print(f"尝试连接到 MySQL 数据库: {DB_CONFIG['database']}...")
            conn = mysql.connector.connect(**DB_CONFIG)
            cursor = conn.cursor()
            print("连接成功。")

            print(f"读取 SQL 文件: {SQL_FILE_PATH}...")
            with open(SQL_FILE_PATH, "r", encoding="utf8") as f:
                sql_text = f.read()

            # --- 拆分 SQL ---
            commands = [cmd.strip() for cmd in sql_text.split(";") if cmd.strip()]
            print(commands)
            print(f"找到 {len(commands)} 条 SQL 命令，开始执行...")

            for idx, command in enumerate(commands, 1):
                # 跳过注释
                upper = command.upper()
                if upper.startswith("--") or upper.startswith("/*"):
                    continue

                try:
                    print(f"[{idx}] 执行: {command[:100]}...")
                    cursor.execute(command)
                except mysql.connector.Error as e:
                    print(f"❌ 命令执行失败（第 {idx} 条）")
                    print(command)
                    raise e

            conn.commit()
            print("\n✅ 所有 SQL 命令执行完成并提交。")

        except mysql.connector.OperationalError as e:
            print(f"\n❌ 连接或操作失败: {e}")
            print("请检查 MySQL 服务、配置和网络。")

        except FileNotFoundError:
            print(f"\n❌ 找不到 SQL 文件 {SQL_FILE_PATH}，请检查路径。")

        except Exception as e:
            print(f"\n❌ 发生意外错误: {e}")
            if conn:
                conn.rollback()
                print("事务已回滚（注意：DDL 无法回滚）")

        finally:
            if cursor:
                cursor.close()
            if conn:
                conn.close()
                print("数据库连接已关闭。")

    execute_sql_file()
```
